```python
natural = df[df['best8']=='natural'].sample(n=20)
modern = df[df['best8']=='modern'].sample(n=20)
minimal = df[df['best8']=='minimal'].sample(n=20)
vintage = df[df['best8']=='vintage'].sample(n=20)
casual = df[df['best8']=='casual']
useful = df[df['best8']=='useful']
individuality = df[df['best8']=='individuality']
romantic = df[df['best8']=='romantic']
```

```python
thanos_final = pd.concat([natural, modern, minimal, vintage, casual, useful, individuality, romantic], axis=0)
thanos_final
```

```python
types = thanos_final.loc[:, 'analysts':'sentinels']
sex = thanos_final.loc[:, '남자':'여자']
age = thanos_final.loc[:, '10대':'60대 이상']
job = thanos_final.loc[:, '대학생':'프리랜서']
y_data= thanos_final.best8
```

```python
y_data.value_counts()
```

```python
y_data.replace(['minimal', 'natural', 'modern', 'vintage', 'casual', 'useful', 'individuality', 'romantic'], 
               [0, 1, 2, 3, 4, 5, 6, 7], inplace=True)
```

```python
from tensorflow.keras.utils import to_categorical

# quality가 3부터 시작하므로 0부터 시작하게끔 
y_data = to_categorical(y_data) 
y_data
```

```python
X_data = pd.concat([types, sex, age, job], axis=1)
X_data
```

```python
# train/test 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_data, 
                                                    y_data, 
                                                    test_size=0.3, 
                                                    shuffle=True, 
                                                    random_state=12)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# 심층 신경망 모델
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout

def build_model(train_data, train_target):
    model = Sequential()
    model.add(Dense(128, activation='tanh', input_dim=X_train.shape[1]))
    model.add(Dropout(0.2))
    model.add(Dense(64, activation='tanh'))
    model.add(Dropout(0.2))
    model.add(Dense(32, activation='tanh'))
    model.add(Dropout(0.2))
    model.add(Dense(y_train.shape[1], activation='softmax'))
    
    model.compile(optimizer='rmsprop',
                  loss='categorical_crossentropy', 
                  metrics=['acc', 'mae'])
    return model

model = build_model(X_train, y_train)
model.summary()
```

```python
# model fit
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss',  patience=10)
history = model.fit(X_train, y_train, batch_size=32, epochs=200,
                    validation_data=(X_test, y_test),
                    callbacks=[early_stopping], 
                    verbose=2)
```

```python
# accuracy
loss, acc, _ = model.evaluate(X_test, y_test)
print('loss>> ', loss)
print('acc>> ', acc)
```

```python
# test 데이터에 대한 예측값 정리
y_pred_proba = model.predict(X_test)
y_pred_proba
```

```python
# 확률을 타겟값으로 변경 => np.argmax
y_pred_label = np.argmax(y_pred_proba, axis=-1)
y_pred_label
```
